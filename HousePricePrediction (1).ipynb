```python
#Data Quality report for Continuous features
numeric_cols = df.select_dtypes(include=['number']).columns
n_rows = df.shape[0]
counts = (n_rows - df[numeric_cols].isna().sum()).to_list()
cont_qr_df = pd.DataFrame(numeric_cols, columns=['Column'])
cont_qr_df['Count'] = counts
perc = (df[numeric_cols].isnull().sum() / n_rows * 100).to_list()
cont_qr_df['% of Missing'] = perc
cont_qr_df['Card.'] = df[numeric_cols].nunique().to_list()
df4 = df[numeric_cols].copy()
for col in numeric_cols:
    cont_qr_df['Min.'] = df4.min().astype('int').to_list()
    cont_qr_df['Q1'] = df4.quantile(0.25).astype('int').to_list()
    cont_qr_df['Median'] = df4.median().astype('int').to_list()
    cont_qr_df['Q3'] = df4.quantile(0.75).astype('int').to_list()
    cont_qr_df['Max.'] = df4.max().astype('int').to_list()
    cont_qr_df['Mean'] = df4.mean().to_list()
    cont_qr_df['Std. Dev.'] = df4.std().to_list()
cont_qr_df
```

```python
#Data Quality report for Categorical features
cat_cols = df.select_dtypes(include=['object']).columns
n_rows = df.shape[0]
counts = (n_rows-df[cat_cols].isna().sum()).to_list()
cat_qr_df = pd.DataFrame(numeric_cols, columns=['Column'])
cat_qr_df = pd.DataFrame(cat_cols, columns=['Column'])
cat_qr_df['Count'] = counts
perc = (df[cat_cols].isna().sum()/n_rows*100).astype('int').to_list()
cat_qr_df['% of Missing'] = perc
df5 = df[cat_cols]
cat_qr_df['Card.'] = df5.nunique().to_list()
count = 0
for i in cat_cols:
    cat_qr_df.loc[count, 'Mode'] = df5[i].value_counts().index[0]
    cat_qr_df.loc[count, 'Mode Freq.'] = df5[i].value_counts()[df5[i].value_counts().index[0]]
    cat_qr_df.loc[count, 'Mode %'] = df5[i].value_counts()[df5[i].value_counts().index[0]]
    cat_qr_df.loc[count, '2nd Mode'] = df5[i].value_counts().index[1]
    cat_qr_df.loc[count, '2nd Mode Freq.'] = 100*df5[i].value_counts()[df5[i].value_counts().index[1]]/n_rows
    cat_qr_df.loc[count, '2nd Mode Perc'] = 100*df5[i].value_counts()[df5[i].value_counts().index[1]]/n_rows
    count = count + 1 
cat_qr_df
```

```python
for feature in numeric_cols:
    df2.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```

```python
for col in numeric_cols:
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    num_outliers = ((df2[col] < lower_bound) | (df2[col] > upper_bound)).sum()
    # Clamp the values (replace outliers with lower/upper bounds)
    df2[col] = df2[col].clip(lower=lower_bound, upper=upper_bound)
    # Count how many values were changed to the lower or upper bound
    
    df3=df2.copy()
    print(f"Column: {col}")
    print(f"Lower Bound: {lower_bound}")
    print(f"Upper Bound: {upper_bound}")
    print(f"Number of outliers clamped: {num_outliers}")
    print("-" * 50)

print("Shape of df after clamping outliers:", df2.shape)
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, r2_score

def test_model(model, X_train=X_train, y_train=y_train):
    cv = KFold(n_splits = 3, shuffle=True, random_state = 45)
    r2 = make_scorer(r2_score)
    r2_val_score = cross_val_score(model, X_train, y_train, cv=cv, scoring = r2)
    score = [r2_val_score.mean()]
    return score
```
